 # Step 1: Connect to the uploaded SQLite database


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("sales_data.db")
```

# Step 2: Write SQL query to get total quantity and revenue per product

```python
query = """
SELECT 
    product, 
    SUM(quantity) AS total_qty, 
    SUM(quantity * price) AS revenue 
FROM sales 
GROUP BY product
"""
```

# Step 3: Load the results into a pandas DataFrame

```python
df = pd.read_sql_query(query, conn)
conn.close()
```

# Step 4: Print the result

```python
print("ðŸ“Š Sales Summary (Total Quantity and Revenue):")
print(df)
```

# Step 5: Create a bar chart for revenue

```python
df.plot(kind='bar', x='product', y='revenue', legend=False, color='teal')
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("sales_chart.png")  # Optional: save the chart
plt.show()
```
